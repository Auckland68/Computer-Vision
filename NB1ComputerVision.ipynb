```python
# Get chunks through slicing
mat[0:3,0:3] # all rows to 3 and all the columns to 3
```

```python
# Reassign values - can work on a copy instead
mat[0:3,0:3] = 0
mat
```

## Images

* Pixels in greyscale images range from 0 - darkest to 255 which is brightest. 

* Colour images - combine Red, Green, Blue (RGB) to create different colours. Each colour has intensity values between 0 to 255. 

* An image will have 3 dimensions or channels, one for each colour. The array is 3 dimensional. The user has to specify which colour each channel is.

```python
# Import PIL
from PIL import Image
```

```python
# Open image with PIL
picture = Image.open("Projects/Images/puppy.jpg")
picture
```

```python
# Get the type
type(picture)
```

```python
# Transform to an array, get shape and view from array form using matplotlib
dog_array = np.array(picture)
print(type(dog_array))
print(dog_array.shape)
plt.imshow(dog_array)
```

```python
# Access a channel (R,G,B) so choose all the rows and columns but just the first channel which is red
dog_red = dog_array.copy()
plt.imshow(dog_red[:,:,0])  # shown in viridis colourmap
print(dog_red[:,:,0].shape) # note the shape for the single channel
```

```python
# Choose as greyscale
red_pic = plt.imshow(dog_red[:,:,0], cmap = "gray")
red_pic
```

Where the pixels are closer to 255 this represents more red. Less red is shown as nearer to zero. The same goes for each of the other channels

```python
# Look at picture for green
green_pic = plt.imshow(dog_red[:,:,1],cmap = "gray")
```

```python
# Look at picture for blue
blue_pic = plt.imshow(dog_red[:,:,2],cmap = "gray")
```

Each picture is slightly different but it is very difficult to tell as they are all represented in grayscale

```python
# Set channel values to zero (R,G,B)
dog_red[:,:,1] = 0  # set all in the green channel to zero
plt.imshow(dog_red) # the only colours left are blue and red (purple)
```

```python
# Set the blue channel to zero
dog_red[:,:,2] = 0
plt.imshow(dog_red) # removed blue and green and so left with red
```
